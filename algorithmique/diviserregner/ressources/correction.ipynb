```python
def inserer(tab:list, j:int) -> None:
    if j >= 0 and tab[j] > tab[j+1]:
        tab[j], tab[j+1] = tab[j+1], tab[j]
        inserer(tab, j-1)



def tri_insertion_rec(tab:list) -> None:
    for i in range(len(tab)):
        inserer(tab, i-1)
```

## Exercice 3
```python
def dichotomie_imp(tab: list, x:int) ->int: 
    debut, fin = 0, len(tab) - 1 
    while debut <= fin: 
        milieu = (debut + fin)//2
        if tab[milieu] == x: 
            return milieu
        elif tab[milieu] < x:
            debut = milieu+1
        else:
            fin = milieu-1 
        return -1
#Version impérative


def dicho_rec(tab: list, x: int, debut:int, fin: int) ->int: 
    if debut <= fin: 
        milieu = (debut + fin)//2
        if tab[milieu] == x:
            return milieu
        elif tab[milieu] < x:
            return dicho_rec(tab, x, milieu + 1, fin)
        else:
            return dicho_rec(tab, x, debut, milieu - 1)
    else:
        return -1

# Version récursive
```
